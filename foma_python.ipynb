```python
%%fst --dot
define Vowel [a|e|i|o|u];
regex Vowel;
```

```python
%%fst --dot
define Vowel [a|e|i|o|u];
regex Vowel+;
```

```python
%%fst --dot
define Consonant [p|b|t|d|k|g];
regex Consonant;
```

```python
%%fst --dot
define CvSyllable Consonant Vowel;
regex CvSyllable;
```

```python
%%fst --dot
define CvcSyllable Consonant Vowel Consonant;
regex CvcSyllable;
```

```python
%%fst --dot
define Syllable CvSyllable | CvcSyllable;
regex Syllable;
```

```python
%%fst --dot
define Word Syllable+;  
regex Word;
```

```python
%%fst --net  
regex Word;
```
